```python
import pandas as pd
```

```python
mydict={
    "iq":[90,100,110,120,80],
    "cgpa":[8,7,6,9,5],
    "salary":[3,4,8,6,3]
}
```

```python
data=pd.DataFrame(mydict)
```

```python
data
```

```python
x=data[["iq","cgpa"]]
y=data.salary
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
fig,ax=plt.subplots(figsize=(10,10))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.rand(100,1)-0.5
y=3*x[:,0]**2+0.05*np.random.randn(100)
```

```python
x
```

```python
y
```

```python
df=pd.DataFrame()
```

```python
df['x']=x.reshape(100)
```

```python
df['y']=y
```

```python
df
```

```python
plt.scatter(df['x'],df['y'])
plt.show()
```

```python
df.head()
```

```python
df['pre1']=df['y'].mean()
```

```python
df
```

```python
df['res1']=df['y']-df['pre1']
```

```python
df
```

```python
plt.scatter(df['x'],df['y'])
plt.plot(df['x'],df['pre1'],color="red")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree1.fit(df['x'].values.reshape(100,1),df['res1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1,filled=True)
plt.show()
```

```python
x_test=np.linspace(-0.5,0.5,500)
```

```python
x_test
```

```python
y_pre=0.265+tree1.predict(x_test.reshape(500,1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(x_test,y_pre,linewidth=2,color="red")
plt.scatter(df['x'],df['y'])
```
