```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=10000, n_features=4, n_informative=2, n_redundant=0,random_state=0, shuffle=False)
```

```python
clf = BaggingClassifier(estimator=SVC(),n_estimators=10, random_state=0).fit(X, y) ## n_estimators means number of model we have traning
```

```python
clf.predict([[0, 0, 0, 0]])
```

```python
clf.score(X,y)
```

```python
clf = BaggingClassifier(estimator=SVC(),n_estimators=50, random_state=0).fit(X, y) ## n_estimators means number of model we have traning
```

```python
clf.score(X,y)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
clf3 = SVC()
```

```python
eclf1 = VotingClassifier(estimators=[
        ('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='hard')
```

```python
eclf1 = eclf1.fit(X, y)
```

```python
eclf1.score(X,y)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.svm import LinearSVC
```

```python

estimators = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('svr',LinearSVC(random_state=42))
                          
]
clf = StackingClassifier(
    estimators=estimators, final_estimator=LogisticRegression()
)
```

```python
clf.fit(X,y)
```

```python
clf.score(X,y)
```
